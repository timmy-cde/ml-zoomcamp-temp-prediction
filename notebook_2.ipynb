Encode cyclical features (month, day of week and hour)

```python
def encode_cyclical(data, col, max_val):
    data[f'{col}_sin'] = np.sin((df[col] * 2 * np.pi) / max_val)
    data[f'{col}_cos'] = np.cos((df[col] * 2 * np.pi) / max_val)

    return data
```

```python
df = encode_cyclical(df, 'day_of_year', 365)
df = encode_cyclical(df, 'month', 12)
df = encode_cyclical(df, 'day_of_week', 7)
df = encode_cyclical(df, 'hour', 24)
```

### Model Training and  Hyperparameter Tuning

Since the purpose of the model to be trained is to predict future temperatures, the models to be used are regressions and not classifications.

There are 2 targets to be predicted which is the temperature at 2m and the apparent temperature, so the models that will be used supports multioutput regression. (More info in sklearn documentation: [Multiclass and multioutput algorithms](https://scikit-learn.org/stable/modules/multiclass.html))

Since this is also a time-dependent data, shuffling datasets is not recommended, and TimeSeriesSplit was used instead of KFold for cross validation to respect the temporal order of observations. 

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_validate, TimeSeriesSplit
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import root_mean_squared_error, r2_score

from xgboost import XGBRegressor

import tensorflow as tf
from tensorflow import keras
```

```python
X_full_train, X_test, y_full_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
X_train, X_val, y_train, y_val = train_test_split(X_full_train, y_full_train, test_size=0.25, shuffle=False)
```

```python
def evaluate(model, y_val, y_pred, cv_only=False):
    if cv_only:
        rmse = np.nan
        r2 = np.nan
    else:
        rmse = root_mean_squared_error(y_val, y_pred)
        r2 = r2_score(y_val, y_pred)
    
    tscv = TimeSeriesSplit(n_splits=10)
    scores = cross_validate(
        model, 
        X_full_train, 
        y_full_train, 
        cv=tscv,
        scoring=('neg_mean_squared_error'),
    )

    score_mean = np.mean(scores['test_score'])
    score_std = np.std(scores['test_score'])

    return (rmse, r2, score_mean, score_std)
```

```python
def print_eval(rmse, r2, score_mean, score_std, cv_only=False):
    if cv_only == False:
        print(f'root_mean_squared_error: {rmse}')
        print(f'r2_score: {r2}')
    
    print('rmse mean: %.3f\nrmse std: %.3f' % (score_mean, score_std))
```

#### Linear Regression Model

```python
model_lr = LinearRegression()
model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_val)
```

```python
plt.figure(figsize=(10, 6))
# plt.plot(range(len(y_test[:, 0])), y_test[:, 0], linestyle='--', label="original")
# plt.plot(range(len(y_test[:, 0])), y_pred[:, 0], label="predicted")
# sns.histplot(y_test[:, 0], alpha=0.5, label="original")
sns.histplot(y_pred[:, 0], alpha=0.5, label="predicted")
plt.legend()
plt.show()
```

```python
rmse, r2, score_mean, score_std = evaluate(model_lr, y_val, y_pred)
print_eval(rmse, r2, score_mean, score_std)
```

#### Ridge Model

```python
scores = {}

for alpha in [0, 0.01, 0.1, 1, 10, 100]:
    model_rd = Ridge(alpha=alpha, random_state=42, max_iter=1000)
    model_rd.fit(X_train, y_train)
    y_pred = model_rd.predict(X_val)
    scores[alpha] = evaluate(model_rd, y_val, y_pred)
```

```python
scores_df = pd.DataFrame(scores, index=['rmse', 'r2', 'score_mean', 'score_std'])
```

#### Random Forest Regression

```python
params = {
    'n_estimators': np.arange(100, 210, 10),
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_leaf': [1, 3, 5, 10, 50],
}

rf = RandomForestRegressor(random_state=42)

tscv_rf = TimeSeriesSplit(n_splits=5)

model_rf = RandomizedSearchCV(
    estimator=rf,
    param_distributions=params,
    cv=tscv_rf,
    scoring=('neg_mean_squared_error'),
    random_state=42
)

model_rf.fit(X_train, y_train)
```

```python
print(f'best score (rmse): {model_rf.best_score_}')
print(f'best parameters: {model_rf.best_params_}')
```
