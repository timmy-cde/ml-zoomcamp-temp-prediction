To reduce complexity and speed up the training process, I will only choose one (1) city out of the 135 cities present in the dataset

```python
df_manila_2020_2023 = df_hourly_2020_2023[df_hourly_2020_2023.city_name == "Manila"]
df_manila_2024 = df_hourly_2024[df_hourly_2024.city_name == "Manila"]
```

Since there are too many null values in 10/23/2024 and 10/24/2024, and the max date is at 10/24/2024, these 2 dates will be removed from the dataset of 2024

```python
df_manila_2024 = df_manila_2024[df_manila_2024.datetime < '2024-10-23']
```

Snow is irrelevant to this country because Philippines is a tropical country. Therefore, `snowfall` and `snow_depth` can be removed.

```python
del df_manila_2020_2023['snowfall']
del df_manila_2020_2023['snow_depth']

del df_manila_2024['snowfall']
del df_manila_2024['snow_depth']
```

Convert `datetime` column to DateTime datatype

```python
df_manila_2020_2023['datetime'] = pd.to_datetime(df_manila_2020_2023['datetime'])
df_manila_2024['datetime'] = pd.to_datetime(df_manila_2024['datetime'])
```

`city_name` is not needed since the data was focused on 1 city only

```python
del df_manila_2020_2023['city_name']
del df_manila_2024['city_name']
```

Combine data of 2020-2023 and 2024

Encode cyclical features (month, day of week and hour)

```python
def encode_cyclical(data, col, max_val):
    data[f'{col}_sin'] = np.sin((df[col] * 2 * np.pi) / max_val)
    data[f'{col}_cos'] = np.cos((df[col] * 2 * np.pi) / max_val)

    return data
```

```python
df = encode_cyclical(df, 'day_of_year', 365)
df = encode_cyclical(df, 'month', 12)
df = encode_cyclical(df, 'day_of_week', 7)
df = encode_cyclical(df, 'hour', 24)
```

```python
grouped_data = df.groupby(['year']).mean().reset_index()

data = grouped_data[['year', 'temperature_2m', 'apparent_temperature']]

# Bar plot setup
x = np.arange(len(data['year']))  
width = 0.30

plt.figure(figsize=(10, 6))

# Bars for temperature_2m
plt.bar(x - width/2, data['temperature_2m'], width, label='temperature_2m', color='blue')

# Bars for apparent_temperature
plt.bar(x + width/2, data['apparent_temperature'], width, label='apparent_temperature', color='orange')

plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Temperature Comparison by Year')
plt.xticks(x, data['year'])  
plt.ylim(25, 35)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
sns.lineplot(data=df, x='year', y='temperature_2m')
plt.title('Yearly Temperature Trends')
plt.xlabel('Date')
plt.xticks(df['year'].unique())  
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
grouped_data = df.groupby(['year', 'month']).mean().reset_index()
data = grouped_data[['year', 'month', 'temperature_2m', 'apparent_temperature']]

for col in ['temperature_2m', 'apparent_temperature']:
    data_pivot = data.pivot(index='month', columns='year', values=col)
    months = data_pivot.index
    years = data_pivot.columns

    x = np.arange(len(months))
    width = 0.15

    plt.figure(figsize=(12, 4))

    for i, year in enumerate(years):
        plt.bar(x + i * width, data_pivot[year], width, label=str(year))

    # Add labels, title, and legend
    plt.xlabel('Month')
    plt.ylabel('Temperature (°C)')
    plt.title(f'{col} Comparison by Month')
    plt.xticks(x + width * (len(years) - 1) / 2, months)
    if col == 'temperature_2m':
        plt.ylim(25, 31)
    else:
        plt.ylim(25, 36)

    plt.legend(title="Year")

    # Show the plot
    plt.tight_layout()
    plt.show()
```

```python
grouped_data = df.groupby(['year', 'day_of_week']).mean().reset_index()
data = grouped_data[['year', 'day_of_week', 'temperature_2m', 'apparent_temperature']]

for col in ['temperature_2m', 'apparent_temperature']:
    data_pivot = data.pivot(index='day_of_week', columns='year', values=col)
    day_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    years = data_pivot.columns

    x = np.arange(len(day_of_week))
    width = 0.15

    plt.figure(figsize=(12, 4))

    for i, year in enumerate(years):
        plt.bar(x + i * width, data_pivot[year], width, label=str(year))

    
    plt.xlabel('Day of Week')
    plt.ylabel('Temperature (°C)')
    plt.title(f'{col} Mean Comparison by Week')
    plt.xticks(x + width * (len(years) - 1) / 2, day_of_week)
    if col == 'temperature_2m':
        plt.ylim(25, 31)
    else:
        plt.ylim(25, 40)

    plt.legend(title="Year")

    
    plt.tight_layout()
    plt.show()
```
